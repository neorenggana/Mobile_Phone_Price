```python
df["Price ($)"] = df["Price ($)"].str.replace('[^0-9]','',regex=True)
df['Price ($)'] = df['Price ($)'].astype(int)
df.describe()
```

```python
df["Price ($)"] = df["Price ($)"].str.replace('[^a-zA-Z0-9]+','',regex=True)
df['Price ($)'] = df['Price ($)'].astype(int)
df.describe()
```
